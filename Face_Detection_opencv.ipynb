FACE DETECTION USING OPENCV 

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
test_image=cv2.imread('C:/Users/sletz/Desktop/Python/Face Detection/Face-Detection-Using-OpenCV/baby.jpg')
testimg_togray=cv2.cvtColor(test_image,cv2.COLOR_BGR2GRAY)
plt.imshow(testimg_togray,cmap='gray')
```

```python
def convertToRGB(imge):
    return cv2.cvtColor(imge,cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_face=cv2.CascadeClassifier('C:/Users/sletz/Desktop/Python/Face Detection/Face-Detection-Using-OpenCV/haarcascade_frontalface_alt.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(testimg_togray, scaleFactor = 1.2, minNeighbors = 5);

# Let us print the no. of faces found
print('Faces found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
     cv2.rectangle(test_image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
#convert image to RGB and show image
rgb_img=convertToRGB(test_image)
plt.imshow(rgb_img)
```

Generalised function for face detection

```python
def detect_faces(cascade, test_image, scaleFactor = 1.1):
    image_copy = test_image.copy()
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)

    faces_rect = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=5)
    
    for (x, y, w, h) in faces_rect:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 15)
        
    return image_copy
```

testing generalised function on a group photograph

```python

test_image2 = cv2.imread('group.jpg')
faces = detect_faces(haar_cascade_face, test_image2)
plt.imshow(convertToRGB(faces))
```
